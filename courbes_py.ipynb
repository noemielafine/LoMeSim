```python
import numpy as np
def load_data(filename):
    data = np.loadtxt(filename)
    x_values = data[:, 0]
    y_values = data[:, 1]
    z_values = data[:, 2]
    return x_values, y_values, z_values
```

```python
import matplotlib.pyplot as plt

files = [
    "pressure_frequence_withmemb_v4.dat",
    "pressure_frequence_withmemb_v4_ext.dat",
]
plt.figure(figsize=(10, 6))
config = ["Sans ext", "Avec ext"]
for i, file in enumerate(files):
    x_values, y_values = load_data(file)
    plt.plot(x_values, y_values, label=config[i])

# Tracer les donnes
plt.xscale('log')
plt.xlabel('Frquence (Hz)')
plt.xlim(20, 20000)
plt.ylabel("SPL (dB)")
plt.title("Spectre de la pression RS100-4 Ohm avec membrane 1")
plt.grid(True, which="both", ls="--")
plt.minorticks_on()
plt.legend()
plt.tick_params(axis='both', which='both', direction='in', top=True, right=True)
plt.savefig('pressure_frequence_witmemb_v4_ext.png')
# Afficher le graphique
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

N = ['10*30', '15*35', '20*40']

F = np.zeros((9, 6))  # Use a NumPy array instead of a list
n = 0

plt.figure(figsize=(10, 6))
for j in range(2):
    for i in range(3):
        FP = np.loadtxt(f"FP_P{j+1}_N{i+1}_withmemb_v4.dat")
        FP.sort()
        for k in range(9):
            F[k, n] = FP[k]  # Correctly assign values to the NumPy array
        n += 1
for i in range(9):
    plt.plot(N[0:3], F[i, 0:3], '-o', label=f'FP{i+1}')

plt.xlabel('Rsolution (Nac*Nst) du maillage')
plt.ylabel("Frquence (Hz)")
plt.title("Premires frquences propres en fonction de la rsolution du maillage - Elements P1")
plt.grid()
plt.legend()
plt.savefig('etude_convergence_P1.png')
plt.show()

plt.figure(figsize=(10, 6))
for i in range(9):
    plt.plot(N, F[i, 3:6], '-o', label=f'FP{i+1}')

plt.xlabel('Rsolution (Nac*Nst) du maillage')
plt.ylabel("Frquence (Hz)")
plt.title("Premires frquences propres en fonction de la rsolution du maillage - Elements P2")
plt.grid()
plt.legend()
plt.savefig('etude_convergence_P2.png')
plt.show()
```
